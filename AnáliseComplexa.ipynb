## Neste caderno vamos poder visualizar algumas funções complexas conhecidas.

#### Parte 1 :  Programação

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000

Bound = 4
```

```python
R = np.linspace(-Bound, Bound, N)
```

```python
X, Y = np.meshgrid(R, 1j*R)

C = X + Y
```

```python
def PlotComplexFunction(F, Title=''):
    
    plt.figure(figsize=(10,10))
    
    plt.title(Title, size=15)
    
    plt.yticks(np.arange(-Bound, Bound+1), [r'{}i'.format(t) for t in np.arange(-Bound, Bound+1)])
    
    plt.axvline(x = 0, c='black', linestyle='--', linewidth=0.5)
    plt.axhline(y = 0, c='black', linestyle='--', linewidth=0.5)
    
    
    FC = F(C)
    
    ABS = np.abs(FC)
    
    q = np.quantile(ABS, 0.999)

    ABS[ABS > q] = q
    
    plt.imshow(ABS, extent=[-Bound, Bound, -Bound, Bound], cmap='BuPu');
    #plt.colorbar(fraction=0.0453);
    
    
    U = np.real(FC).flatten()
    V = np.imag(FC).flatten()
    
    Norms = np.sqrt(U**2 + V**2)
    
    U = U / Norms
    V = V / Norms
    
    X = np.real(C).flatten()
    Y = np.imag(C).flatten()
    
    QuiverX = X.reshape(N, N)[::40, ::40].flatten()
    QuiverY = Y.reshape(N, N)[::40, ::40].flatten()
    QuiverU = U.reshape(N, N)[::40, ::40].flatten()
    QuiverV = V.reshape(N, N)[::40, ::40].flatten()

    plt.quiver(QuiverX, QuiverY, QuiverU, QuiverV);
```

```python
@np.vectorize
def F1(z):

    return 1
```

```python
@np.vectorize
def F13(z):

    from scipy.special import gamma
    
    return gamma(z)
```
